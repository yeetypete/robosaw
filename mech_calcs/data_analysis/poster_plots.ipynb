```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data_path = './data/'
save_path = './plots/'
plot_style = 'robosaw.mplstyle'
plt.style.use(plot_style)

```

```python
pid_results = pd.read_csv(data_path + "robosaw_pid_2x4.csv", encoding='utf-8', header=0)
pid_results = pid_results.truncate(before=0, after=40)
pid_results['time'] -= pid_results['time'][0]
```

```python
fig1, ax1 = plt.subplots(figsize=(6,4))
ax1.plot(pid_results['time'], pid_results['pid_error'], lw=3, color='#F5613A')
ax1.set_xlabel(r"Time, $s$ (seconds)")
ax1.set_ylabel(r"Error, $e(t)$ (pixels)")
ax1.set_title(r"RoboSaw PID Controller Error with 2x4")
fig1.show()
fig1.savefig(save_path + "robosaw_pid_2x4.png", bbox_inches='tight', dpi=200)
```
